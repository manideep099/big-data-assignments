```python
# Part 1: String operations

# Input string from user
s = input("Enter a string: ")

# Convert string to list of characters
char_list = list(s)

# Delete at least 2 characters (removing first and last characters)
if len(char_list) >= 2:
    del char_list[0]
    del char_list[-1]

# Reverse the resultant list
char_list.reverse()

# Convert back to string and print
result = "".join(char_list)
print("Resultant string:", result)


# Part 2: Arithmetic operations

# Input two numbers
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

# Perform arithmetic operations
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)

# Division (check to avoid division by zero)
if b != 0:
    print("Division:", a / b)
else:
    print("Division: Not possible (division by zero)")
```

```python
# Input class score
score = int(input("Enter the class score: "))

# Determine the letter grade
if score >= 90 and score <= 100:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```
